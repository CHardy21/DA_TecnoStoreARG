```python
import pandas as pd

def load_data():
    df = pd.read_parquet("data/analytical_layer/tecnoStore_modelo_estrella.parquet", engine="fastparquet")
    return df

df = load_data()
df.head(3)
```

```python
print(df["Provincia"].unique())
```

```python
df_filtrado = df
ventas_nominal = df_filtrado['monto_venta_ars_nominal'].sum()
ventas_online = df_filtrado.loc[df_filtrado["Canal_Venta"] == "Online", "monto_venta_ars_nominal"].sum()
#ventas_total = df_filtrado["ventas"].sum()
participacion_online = (ventas_online / ventas_nominal) * 100
print("ventas totales: ", ventas_nominal)
print("ventas online:  ", ventas_online)
print("participacion:  ", participacion_online)
```
